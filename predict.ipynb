```python
import pandas_datareader as pdr
import yfinance as yf
import pandas as pd
from datetime import datetime
```

```python
# Fetch data for TSLA from Yahoo Finance using yfinance
tsla_data = yf.download('TSLA')
tsla_data
```

```python
tsla_data.plot()
```

```python
tsla_data['High'].plot(xlim=['2019-04-01','2024-09-01'],figsize=(12,4))
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
index=tsla_data.loc['2020-01-01': '2021-09-01'].index
share_open=tsla_data.loc['2020-01-01':'2021-09-01']['Open']
```

```python
figure,axis=plt.subplots()
plt.tight_layout()
figure.autofmt_xdate()
#preventing the overlap
axis.plot(index, share_open)
```

```python
df_tsla=tsla_data.reset_index()
```

```python
df_tsla.info()
```

```python
df_tsla
```

```python
df_tesla=df_tsla.set_index('Date',drop=True)
```

```python
df_tesla
```

## time resampling

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').min()['Open'].plot()
```

```python
##quaterly start frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_tesla.resample(rule='QS').max()
```

```python
##Business End Frequency
##https://towardsdatascience.com/resample-function-of-pandas-79b17ec82a78
df_tesla.resample(rule='BA').max()['High'].plot()
```

```python
df_tesla.resample(rule='BA').max()['High']
```

```python
df_tesla.resample(rule='BQS').max()['High']
```

## time plotting

```python
##plotting
df_tesla['Open'].resample(rule='BA').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max() .plot(kind='bar',figsize= (30,6))
```

## rolling

```python
df_tesla['High'].rolling(11).mean().head(20)
# measures the mean in the window sixe of 11 here so first 11 items are nan
# can do mean, min, max anything
```

```python
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(30) .mean()
```
